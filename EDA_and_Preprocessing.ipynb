# Data Preprocessing and Exploratory Data Analysis (EDA)

## Project: Global Meteorite Landings Visualization System
This notebook covers the data extraction, preprocessing, and exploratory data analysis of the Global Meteorite Landings dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')
```

### 1. Data Loading
We load the synthetic/cleaned dataset simulating the Global Meteorite Landings.

```python
df = pd.read_csv('data/cleaned_meteorites.csv')
df.head()
```

### 2. Data Overview & Preprocessing
Check for missing values, correct data types, and bounds.

```python
print("Dataset Information:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())
```

The dataset is already cleaned from our extraction script. It ensures that `lat`, `lng`, `mass`, and `year` are valid floats and integers.

### 3. Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['year'], bins=50, kde=True)
plt.title('Distribution of Meteorite Landings over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='recclass', data=df, order=df['recclass'].value_counts().index)
plt.title('Meteorite Landings by Class')
plt.xlabel('Count')
plt.ylabel('Meteorite Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='lng', y='lat', size='mass', sizes=(10, 200), hue='recclass', data=df, alpha=0.6)
plt.title('Geographical Distribution of Meteorite Landings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Conclusion
The dataset is structured properly and displays clear patterns that are well-suited for an interactive D3.js dashboard, involving temporal, categorical, and geographical dimensions.
